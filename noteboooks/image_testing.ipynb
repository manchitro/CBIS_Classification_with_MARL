```python
import cv2
from PIL import Image
import numpy as np
import torchvision.transforms as tr
from matplotlib import pyplot as plt 

f = open('cbis/jpeg/1.3.6.1.4.1.9590.100.1.2.281397494612871934937455783843630775495/1-195.jpg', 'rb')
img = Image.open(f)
tran = tr.ToTensor()
img = tran(img)
fig = plt.figure()
plt.imshow(img.permute(1, 2, 0), cmap="gray")
plt.plot()
```
